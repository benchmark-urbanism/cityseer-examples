# Metric distance network centrality

Calculate metric distance centralities from a `geopandas` `GeoDataFrame`.


```python
import geopandas as gpd
import matplotlib.pyplot as plt
from cityseer.metrics import networks
from cityseer.tools import graphs, io
```

Prepare the network as shown in other examples. Working with the dual graph is recommended.


```python
streets_gpd = gpd.read_file("data/madrid_streets/street_network.gpkg")
streets_gpd = streets_gpd.explode(reset_index=True)
G = io.nx_from_generic_geopandas(streets_gpd)
```

Load the study area boundary and set `live=True` for nodes inside the boundary.
Nodes outside the boundary act as a buffer to prevent edge rolloff — they are used for routing
but metrics are not computed for them. See the [live nodes example](../live_nodes.ipynb) for more details.


```python
from shapely import geometry

bounds_gpd = gpd.read_file("data/madrid_bounds/madrid_bounds.gpkg")
boundary_poly = bounds_gpd.geometry.iloc[0]
for node_idx, node_data in G.nodes(data=True):
    node_pnt = geometry.Point(node_data["x"], node_data["y"])
    if node_pnt.intersects(boundary_poly):
        G.nodes[node_idx]["live"] = True
    else:
        G.nodes[node_idx]["live"] = False
```

```python
G_dual = graphs.nx_to_dual(G)
```

Use [`network_structure_from_nx`](https://cityseer.benchmarkurbanism.com/tools/io#network-structure-from-nx) from the `cityseer` package's `io` module to prepare the `GeoDataFrames` and `NetworkStructure`.


```python
# prepare the data structures
nodes_gdf, _edges_gdf, network_structure = io.network_structure_from_nx(
    G_dual,
)
```

Use the [`node_centrality_shortest`](https://cityseer.benchmarkurbanism.com/metrics/networks#node-centrality-shortest) function from the `cityseer` package's `networks` module to calculate shortest metric distance centralities. The function requires a `NetworkStructure` and nodes `GeoDataFrame` prepared with the `network_structure_from_nx` function in the previous step.

The function can calculate centralities for numerous distances at once via the `distances` parameter, which accepts a list of distances.

The function returns the nodes `GeoDataFrame` with the outputs of the centralities added as columns. The columns are named `cc_{centrality}_{distance}`. Standard `geopandas` functionality can be used to explore, visualise, or save the results. See the [documentation](https://cityseer.benchmarkurbanism.com/metrics/networks#node-centrality-shortest) for more information on the available centrality formulations.


```python
distances = [500, 2000]
nodes_gdf = networks.node_centrality_shortest(
    network_structure=network_structure,
    nodes_gdf=nodes_gdf,
    distances=distances,
)
nodes_gdf.head()
```

Alternatively, you can define the distance thresholds using a list of `minutes` instead.


```python
nodes_gdf = networks.node_centrality_shortest(
    network_structure=network_structure,
    nodes_gdf=nodes_gdf,
    minutes=[15],
)
```

The function will map the minutes values into the equivalent distances, which are reported in the logged output.


```python
nodes_gdf.columns
```

As per the function logging outputs, 15 minutes has been mapped to 1200m at default `speed_m_s`, so the corresponding outputs can be visualised using the 1200m columns. Use the configurable `speed_m_s` parameter to set a custom metres per second walking speed.


```python
fig, ax = plt.subplots(1, 1, figsize=(8, 6), facecolor="#1d1d1d")
nodes_gdf.plot(
    column="cc_harmonic_1200",
    cmap="magma",
    legend=False,
    ax=ax,
)
ax.set_xlim(438500, 438500 + 3500)
ax.set_ylim(4472500, 4472500 + 3500)
ax.axis(False)
```

For spatial-impedance weighted forms of centralities (beta variants), you can specify the `beta` parameter explicitly. These will otherwise be extrapolated automatically from the `distances` or `minutes` parameters. See the [documentation](https://cityseer.benchmarkurbanism.com/rustalgos/rustalgos#betas-from-distances) for more information on how spatial impedances are converted to distance thresholds.


```python
nodes_gdf = networks.node_centrality_shortest(
    network_structure=network_structure,
    nodes_gdf=nodes_gdf,
    betas=[0.01],
)
```

```python
nodes_gdf.columns
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 6), facecolor="#1d1d1d")
nodes_gdf.plot(
    column="cc_beta_400",
    cmap="magma",
    legend=False,
    ax=ax,
)
ax.set_xlim(438500, 438500 + 3500)
ax.set_ylim(4472500, 4472500 + 3500)
ax.axis(False)
```

## Sampled centrality for larger distances


For larger distance thresholds, the computational cost increases substantially.
The `node_centrality_shortest` function supports a `sample=True` parameter that uses a sampling strategy
to compute centralities more efficiently at larger scales while maintaining a relatively high degree of rank
performance. Rather use non-sampled when generating publication-quality results.


```python
distances = [500, 10000]
nodes_gdf = networks.node_centrality_shortest(
    network_structure=network_structure,
    nodes_gdf=nodes_gdf,
    distances=distances,
    sample=True,
)
nodes_gdf.columns
```

For shorter distances where the number of reachable nodes is small, full computation is used.
For larger distances, the function applies sampling to achieve the target accuracy efficiently.


```python
fig, ax = plt.subplots(1, 1, figsize=(8, 6), facecolor="#1d1d1d")
nodes_gdf.plot(
    column="cc_betweenness_10000",
    cmap="magma",
    legend=False,
    ax=ax,
)
ax.set_xlim(438500, 438500 + 3500)
ax.set_ylim(4472500, 4472500 + 3500)
ax.axis(False)
```

## Tolerance for near-shortest paths


The `tolerance` parameter allows betweenness to count paths that are within a given percentage of the shortest path.
For example, `tolerance=2.0` treats paths within 2% of the shortest as equally valid.
This can produce more realistic betweenness distributions by accounting for near-optimal route choices.


```python
distances = [5000]
nodes_gdf = networks.node_centrality_shortest(
    network_structure=network_structure,
    nodes_gdf=nodes_gdf,
    distances=distances,
    sample=True,
    tolerance=0.5,  # percent <= 1% recommended
)
nodes_gdf.columns
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 6), facecolor="#1d1d1d")
nodes_gdf.plot(
    column="cc_betweenness_5000",
    cmap="magma",
    legend=False,
    ax=ax,
)
ax.set_xlim(438500, 438500 + 3500)
ax.set_ylim(4472500, 4472500 + 3500)
ax.axis(False)
```
